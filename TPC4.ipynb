```python
# -*- coding: utf-8 -*-
"""
Created on Tue Oct 26 21:42:05 2021

@author: MV
"""
from random import randint
jog1=True
jogo=True
Fração=[]
FraçãoResposta=[]
ListaFrações=[]
num=0
denom=0
while jogo==True:
    
    def criar_fração(num,denom):
        return(num,denom)
    
    def simplificar_fração(num,denom):
        def mdc(num,denom):
            if num<denom:
                return mdc(denom,num)
            elif num%denom==0:
                return denom
            else:
                return mdc(num,num%denom)
        mdc=mdc(num,denom)
        num=num/mdc
        denom=denom/mdc
        return(num,denom)
    
    def somar(num,denom,num2,denom2):
        if denom==denom2:
            num3=num+num2
            denom3=denom
        else:
            num3=num*denom2+num2*denom
            denom3=denom*denom2
        return(num3,denom3)
    
    def subtrair(num,denom,num2,denom2):
        if denom==denom2:
            denom3=denom
            num3=num-num2
        else:
            num3=num*denom2-num2*denom
            denom3=denom*denom2
        return(num3,denom3)
    
    def multiplicar(num,denom,num2,denom2):
        num3=num*num2
        denom3=denom*denom2  
        return(num3,denom3)
    
    def dividir(num,denom,num2,denom2):
        num3=num*denom2
        denom3=num2*denom
        return(num3,denom3)
    
    def gerar_lista(N):
        Lnum=[]
        Ldenom=[]
        for i in range(N):
            elem=randint(0,100)
            Lnum.append(elem)
            elem=randint(1,100)
            Ldenom.append(elem)
        return[Lnum,Ldenom]
    
    def soma_lista(lista):
        num=0
        denom=1
        for i in range(len(lista[0])):
            (num,denom) = somar(num, denom, lista[0][i], lista[1][i])
        return(num,denom)
        
    def maior_fração(ListaFrações):
        Lvalores=[]
        for i in range(len(ListaFrações[0])):
            valores=ListaFrações[0][i]/ListaFrações[1][i]
            Lvalores.append(valores)
        M=max(Lvalores)
        IndiceM=Lvalores.index(M)
        NumM=ListaFrações[0][IndiceM]
        DenomM=ListaFrações[1][IndiceM]
        return(NumM,DenomM)
    
    
    def menu(jog1,FraçãoResposta):
        print("""Guião de listas de inteiros:
                      (0)-Sair
                      (1)-Criar fração
                      (2)-Simplificar fração
                      (3)-Somar
                      (4)-Subtrair
                      (5)-Multiplicar
                      (6)-Dividir
                      (7)-Gerar lista de N frações
                      (8)-Somar lista
                      (9)-Maior fração""")
        R=input("Escolha uma opção:\n")
        global Fração
        if R=="1":
            global num
            num=int(input("Numerador: "))
            global denom
            denom=int(input("Denominador: "))
            Fração.clear()
            Fração=(criar_fração(num,denom)) 
            print("A sua fração é", str(Fração[0])+"/"+str(Fração[1]))
            
        elif R=="7":
            N=int(input("Quantas frações deseja criar?\n"))
            global ListaFrações
            ListaFrações.clear()
            ListaFrações=(gerar_lista(N))
            print("As suas frações são:")
            for i in range(N):
                print(str(ListaFrações[0][i])+"/"+str(ListaFrações[1][i])+"\n")
                
        elif jog1==False and R=="2":
            Fração=(simplificar_fração(num,denom))
            print("A sua fração é", str(int(Fração[0]))+"/"+str(int(Fração[1])))
            
        elif jog1==False and R in ["3","4","5","6"]:
            num2=int(input("Segundo numerador: "))
            denom2=int(input("Segundo denominador: "))                 
            if R=="3":
                Fração=somar(num,denom,num2,denom2)
                print("O resultado da soma das duas frações é",str(Fração[0])+"/"+str(Fração[1]))
            elif R=="4":
                Fração=subtrair(num,denom,num2,denom2)
                print("O resultado da subtração das duas frações é",str(Fração[0])+"/"+str(Fração[1]))
            elif R=="5":
                Fração=multiplicar(num,denom,num2,denom2)
                print("O resultado da multiplicação das duas frações é",str(Fração[0])+"/"+str(Fração[1]))
            elif R=="6":
                Fração=dividir(num,denom,num2,denom2)
                print("O resultado da divisão das duas frações é",str(Fração[0])+"/"+str(Fração[1]))
                      
        elif jog1==False and R in ["8","9"]:   
            if R=="8":
                (num,denom)=soma_lista(ListaFrações)
                FraçãoResposta=simplificar_fração(num,denom)
                print("O resultado do somatório das frações é",str(int(FraçãoResposta[0]))+"/"+str(int(FraçãoResposta[1])))
            elif R=="9":
                FraçãoResposta=maior_fração(ListaFrações)
                print("O maior valor das frações é",str(FraçãoResposta[0])+"/"+str(FraçãoResposta[1]))
        elif R=="0":
            global jogo
            jogo=False
            print("Terminou o programa\n")
        else:
            print("Valor inválido!")
    menu(jog1,FraçãoResposta)
    jog1=False
```
